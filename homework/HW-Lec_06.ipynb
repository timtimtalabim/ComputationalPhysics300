```python
#One of the best ways to deal with noisy data is through the use of the scipy method uniform_filter1d.
#This function allows you to create data points based on the averages of a given data set.
#You can control the size of the new data set created through the filtering, allowing you to pick and choose
#The most suitable averaged out graph

from scipy.ndimage.filters import uniform_filter1d
import matplotlib.pyplot as plt
import numpy as np

#define the function
f = lambda x: x * np.sin(x)  

#define the paramters for the plot
a,b = 0,3
npoints = 100

x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5

plt.plot(x,y, label = "Original")
for i in range(3, 100, 10):
  newY = uniform_filter1d(y, size=i)
  plt.plot(x, newY, label=f"{i}")
    
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate

def heatCap(tMin, tMax):
    
    f = lambda x: (x**4 * np.exp(x)) / (np.exp(x) - 1)**2

    v = 1000
    rho = 6.022 * 10**28
    thetaD = 428
    kb = 1.38 * 10**-23
    
    equation = 9 * v * rho * kb

    x = np.linspace(tMin, tMax, (tMax-tMin))
    y = np.linspace(tMin, tMax, (tMax-tMin))
    for t in range(tMin, tMax):
        y[t-tMin] = equation * (t/rho)**3 * integrate.quad(f, 0 , thetaD/t)[0]
    #print(integrate.quad(f, 0 , thetaD/t)[0])
    #print(y)
        

    plt.plot(x, y)
    plt.show()
    
    
```

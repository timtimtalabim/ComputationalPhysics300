```python
import matplotlib.pyplot as plt
import numpy as np
from math import sin
from random import random

def f(x):
    return np.sqrt(1-(x-1)**2)

def f1(x):
    return 2-np.sqrt(4-x**2)
 
def MC(N):
    count = 0
    for i in range(N):
        x = 2*random()
        y = random()
        if f1(x)<y<f(x):
            count += 1
    I = 2*count/N
    return I

def variance_creation(N):
    
    integrals = np.zeros(100)
    
    for i in range (0, 100):
        integrals[i] = MC(N)
    
    mean = sum(integrals) / 100
    
    deviations = [(x - mean) ** 2 for x in integrals]
    
    variance = sum(deviations) / 100
    return variance
    

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')

x1 = np.linspace(0,2,100)
y1 = np.sqrt(1-(x1-1)**2)
ax1.plot(x1,y1,'k')

y2 = 2-np.sqrt(4-x1**2)
ax1.plot(x1,y2,'k')

ax1.set_xlim(0,2)
ax1.set_ylim(0,2)
ax1.fill_between(x1, y1, y2, where=y1>y2, facecolor='green')

plt.show()

for i in [100, 1000, 10000, 100000]:
   print(i, MC(i), variance_creation(i))
```

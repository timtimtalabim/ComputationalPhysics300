```python
%matplotlib inline  

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=[12, 8])
#define the function
f = lambda x: x**x

#define the paramters for the plot
a,b = 0,3
npoints = 100
order = 3

np.random.seed(0)

x = np.linspace(a,b,npoints)
y = f(x) + np.random.rand(npoints) - 0.5  # with random noise

for i in range(5,16,2):
    
    fit = np.polyfit(x, y, i, full=True)
    y_p = np.polyval(fit[0], x)
    plt.plot(x,y_p, label='polyfit'+str(i)+ "err :"+str(fit[1]))

plt.plot(x,f(x), label='ideal function')
plt.plot(x,y, 'o', label='noisy data')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.ylim(0,5)
plt.legend(fontsize=12)
plt.title('fitted coefficients' + str(fit[0]) + '\n residual: ' + str(fit[1]))
plt.show()
#plt.text(2, 2, fit)
```

```python
#For this function it looks like all orders for this functions seems to fit well
#but if you zoom out from 0,3 and increase it to 0,5 you will see that the order of the function will matter
```
